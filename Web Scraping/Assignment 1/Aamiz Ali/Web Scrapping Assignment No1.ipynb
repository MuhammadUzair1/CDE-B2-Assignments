```python
import requests
from bs4 import BeautifulSoup
import re
import json
import csv
import time
```

```python
url = 'https://www.thriftbooks.com/api/browse/Search'
```

```python
headers = {
    'Content-Type': 'application/json',
    'Accept': 'application/json'
}

# Open CSV file once
with open('thrift_books_data.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    # Write header row
    writer.writerow(['Title', 'Condition', 'Buy Price', 'Author', 'Publisher', 'ISBN'])

    # Loop over 200 pages
    for page_num in range(1, 201):
        payload = {
            "currentPage": page_num,
            "displayType": 1,
            "isCollectible": True,
            "isInStock": True,
            "isList": True,
            "itemsPerPage": "50",
            "resultsPerPage": "50",
            "page": page_num,
            "searchTerms": [""],
            "sortBy": "mostPopular",
            "sortDirection": "desc",
            "tags": [12529]
        }

        try:
            response = requests.post(url, headers=headers, json=payload)
            response.raise_for_status()
            data = response.json()

            # Extract works (books)
            works = data.get('works', [])

            for book in works:
                title = book.get('title', '')
                condition = book.get('buyNowCondition', '')
                buy_price = book.get('buyNowPrice', '')
                author = book.get('authors', [{}])[0].get('authorName', '') if book.get('authors') else ''
                publisher = book.get('publisher', '')
                isbn = book.get('iSBN', '')

                writer.writerow([title, condition, buy_price, author, publisher, isbn])

            print(f"Page {page_num} processed successfully.")

            # Optional delay to avoid rate limiting
            time.sleep(0.5)

        except Exception as e:
            print(f"Error on page {page_num}: {e}")

print("Data saved to thrift_books_data.csv successfully!")
```
