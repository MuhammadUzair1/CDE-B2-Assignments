```python
import requests
from bs4 import BeautifulSoup
```

```python
base_url = 'https://www.thriftbooks.com/browse/?b.search=backend&b.s=mostPopular-desc&b.p={page}&b.pp=50&b.oos&b.tile'
```

```python
PAGES = 5

for page in range(1, PAGES + 1):
    url = base_url.format(page=page)
    print(f'\n  Page: {page}...')

    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    bookItems = soup.find_all('div', class_='AllEditionsItem-tile Recipe-default')
    if not bookItems:
        print('No more books found.')
        break

    for book in bookItems:
        title_div = book.find('div', class_='AllEditionsItem-tileTitle')
        title = title_div.find('a').text.strip() if title_div else 'No title'

        price_div = book.find('div', class_='SearchResultListItem-dollarAmount')
        price = price_div.text.strip() if price_div else 'Price not available'

        author_div = book.find('div', class_='SearchResultListItem-bottomSpacing')
        author = author_div.find('a').text.strip() if author_div and author_div.find('a') else 'Author not available'

        image_div = book.find('div', class_='SearchResultTileItem-photo')
        img_tag = image_div.find('img') if image_div else None

        if img_tag:
         image_url = img_tag.get('src') or img_tag.get('data-src', 'Image not available')
        else:
         image_url = 'Image not available'

        print(f'Title: {title}')
        print(f'Author: {author}')
        print(f'Price: {price}')
        print(f'Image: {image_url}\n')
```
