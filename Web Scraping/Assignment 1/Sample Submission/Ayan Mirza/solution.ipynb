```python
import requests
from bs4 import BeautifulSoup
import json 
import re
```

```python
response = requests.get("https://www.thriftbooks.com/browse/?13362col#b.s=mostPopular-desc&b.p=1&b.pp=50&b.col&b.f.t%5B%5D=13362&b.list")
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
scripts = soup.find_all('script')
script_content = str(scripts[12].string)

match = re.search(
    r'window\.searchStoreV2\s*=\s*({.*?});', 
    script_content, 
    re.DOTALL
)

if match:
    raw = match.group(1)
    data_json = json.loads(raw)
    data = data_json.get('works', [])
    
data
```

```python
with open('thriftBooks.csv', 'w') as file:
    file.write('Name,Author,Publisher,LowPrice,HighPrice,ISBN,Qulity Description,\n')

for item in data:
    name = item.get('title')
    author = item.get('authors')[0].get('authorName')
    publisher = item.get('publisher')
    lowPrice = item.get('lowPrice')
    highPrice = item.get('highPrice')
    isbn = item.get('iSBN')
    qualityDescription = item.get('qualityDescription')

    with open('thriftBooks.csv', 'a') as file:
        file.write(f'{name},{author},{publisher},{lowPrice},{highPrice},{isbn},{qualityDescription}\n')
```
